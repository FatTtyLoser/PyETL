### Request

```python
# å° API é€²è¡Œ GET è«‹æ±‚
req = request.Request(url=url, headers=headers)
res = request.urlopen(req)
html = res.read().decode('utf-8')

# å°‡ html.txt ç”¨ json è§£æå½¢æˆ list
jsonData = json.loads(html)
jsonData
```

```python
# é€é loop æŠŠ list æ‹†é–‹é€ä¸€é‡å° json çš„æ¨™ç±¤åšé¸å–
# title = æ¯ç¯‡æ–‡ç« çš„æ¨™é¡Œï¼ŒmediaMeta = æ¯ç¯‡æ–‡ç« çš„ img æ ¼å¼ç‚º listï¼Œid = æ¯ç¯‡æ–‡ç« çš„ url 
for articleDict in jsonData:
    title = articleDict['title']
    articleUrl = 'https://www.dcard.tw/f/photography/p/' + str(articleDict['id'])
    print(title)
    print(articleUrl)
    # åˆ©ç”¨è¿´åœˆæŠŠæ¯ç¯‡æ–‡ç« çš„ img ä¸€å¼µä¸€å¼µå„²å­˜
    for imgs in articleDict['media']:
        print('\t' + imgs['url'])
        time.sleep(2)
        imagePath = './dcardPhoto/{}_{}'.format(title, imgs['url'].split('/')[-1])
        imgContent = requests.get(imgs['url'], headers=headers).content
        with open(imagePath, 'wb') as f:
            f.write(imgContent)
    print()
```

### Import module

```python
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
from urllib import request
import pandas as pd
import random
import time
import json
import ssl
import re
import os
```

### First request

```python
# è¨­å®šæˆ‘å€‘åœ¨ç¨‹å¼ä¹‹ä¸­ä½¿ç”¨ time.sleep ç­‰å¾…çš„æ™‚é–“ï¼Œè®“æ™‚é–“è®Šæ•¸å¯è¨­å®šåˆäº‚æ•¸æŠ½å–
seconds = [ç§’æ•¸1, ç§’æ•¸2, ç§’æ•¸3, ç§’æ•¸4, ç§’æ•¸5......]
delay = random.choice(seconds)

# è¨­å®šå„²å­˜è·¯å¾‘
path = r'.\\dcardtalk'
if not os.path.exists(path):
    os.mkdir(path)

# Dcard å…¬é–‹ API URL
api = 'https://www.dcard.tw/service/api/v2'
# è¨ªå• talk ç‰ˆå‰äº”ç¯‡ç†±é–€æ–‡ç« 
url = api + '/forums/talk/posts?popular=true&limit=5'
# fack_useragent äº‚æ•¸æä¾› headers æé«˜éš±è”½æ€§
headers = {
    'User-Agent' : UserAgent().random
}
req = request.Request(url = url , headers = headers)
res = request.urlopen(req)
html = res.read().decode('utf-8')
rejson = json.loads(html)
rejson
# è¨ªå•äº†ä¸€æ¬¡ä¼ºæœå™¨æ‰€ä»¥é€²è¡Œç­‰å¾…
time.sleep(delay)
```

### First loop

```python
# ç¬¬ä¸€å€‹è¿´åœˆé€ç­†è§£ææ¯ä¸€ç¯‡æ–‡ç« è³‡è¨Š
for articleDict in rejson:
    # å…ˆå»ºç«‹æ–‡ç« å…§å®¹è®Šæ•¸
    articleTxt = ''
    # å–å¾—æ¯ä¸€ç¯‡æ–‡ç« çš„ title ä½œç‚ºå„²å­˜æª”å
    title = articleDict['title']
    # Win10 æª”åæœ‰ç‰¹æ®Šå­—å…ƒé™åˆ¶ï¼Œé€é python æ­£å‰‡åŒ–å°‡ title ç‰¹æ®Šå­—å…ƒåˆªé™¤
    if ':' and '/' and '*' and '"' and '>' and '<' and '|' and '?' and 'ï¼š' and ' ' in title:
        title = re.sub("\<|\'|\?|\ |\>|\*|\:|\"|\ï¼Ÿ|\ï¼š|\||\/|\//","",title)
    else:
        pass
    # å–å¾—æ¯ä¸€ç¯‡æ–‡ç« çš„ ID ä½œç‚ºè¨ªå•å–®ç¯‡æ–‡ç« çš„ API åƒæ•¸
    id = articleDict['id']
    # format id è¨ªå•å–®ç¯‡æ–‡ç« 
    cp_articleUrl = 'https://www.dcard.tw/service/api/v2/posts/{}'.format(id)
    cp_req = request.Request(url = cp_articleUrl , headers = headers)
    cp_res = request.urlopen(cp_req)
    cp_html = cp_res.read().decode('utf-8')
    cp_jsonData = json.loads(cp_html)
    # åŠ å…¥å–®ç¯‡æ–‡ç« çš„å…§å®¹ content 
    articleTxt += cp_jsonData['content'] + '\n=============\n'
    # åˆè¨ªå•ä¸€æ¬¡ä¼ºæœå™¨æ‰€ä»¥é€²è¡Œç­‰å¾…
    time.sleep(delay)
    # å…ˆå»ºç«‹è¨˜æ•¸ç•™è¨€çš„ index
    index = 0
    # ç•™è¨€æ¨“å±¤çš„åƒæ•¸è¨­å®š
    after = 0
    # é€é API è«‹æ±‚å–®ç¯‡æ–‡ç« çš„ç•™è¨€å€ after è¡¨ç¤ºå¾å“ªä¸€å€‹æ¨“å±¤å¾€ä¸‹
    url = 'https://www.dcard.tw/service/api/v2/posts/{}/comments?limit=30&after={}'.format(id, after)
    req = request.Request(url = url , headers = headers)
    res = request.urlopen(req)
    html = res.read().decode('utf-8')
    # æ­¤æ™‚çš„ jsonData æ‡‰è©²æ˜¯å–®ç¯‡æ–‡ç« çš„æ¯ä¸€å€‹ç•™è¨€æ¨“å±¤çš„å…§å®¹
    jsonData = json.loads(html)
    # åˆè¨ªå•ä¸€æ¬¡ä¼ºæœå™¨æ‰€ä»¥é€²è¡Œç­‰å¾…
    time.sleep(delay)
```

### Second loop

```python
    # ç¬¬äºŒå€‹è¿´åœˆè§£ææ¯ä¸€æ¨“å±¤çš„æ¯ä¸€å€‹ç•™è¨€å…§å®¹
    while True:
        # è‹¥æˆ‘å€‘ä¸Šä¸€å±¤è¿´åœˆè£¡é¢é¦–æ¬¡è«‹æ±‚ç•™è¨€å€å–å¾—çš„ç•™è¨€ä¸ç‚ºé›¶ï¼Œé‚£éº¼æˆ‘å€‘ç¢ºå®šæˆ‘å€‘æœ‰éœ€è¦çˆ¬å–çš„ç•™è¨€ï¼Œå› æ­¤é€²è¡Œçˆ¬å–
        if len(jsonData) != 0:
            # å¾ç¬¬0æ¨“é–‹å§‹çˆ¬å–
            url = 'https://www.dcard.tw/service/api/v2/posts/{}/comments?limit=30&after={}'.format(id, after)  
            req = request.Request(url = url , headers = headers)
            res = request.urlopen(req)
            html = res.read().decode('utf-8')
            jsonData = json.loads(html)
#             print(jsonData)
            print(title+str(len(jsonData)))
            # ä½¿ç”¨ try loop ä¾‹å¤–è™•ç†æˆ‘å€‘é‡åˆ°åˆªé™¤ç•™è¨€çš„æ¨“å±¤
            try:
                # å¦‚æœç•™è¨€æ²’è¢«åˆªé™¤å‰‡åŸ·è¡Œä»¥ä¸‹
                for articleCon in jsonData:
    #                 print(articleCon['content'])
                    index += 1
                    con = articleCon['content']
                    # å°‡æ›è¡Œç¬¦è™Ÿåˆªé™¤
                    con = re.sub("\n","",con)
                    # æ•´ç†æ¯ç¯‡ç•™è¨€æ ¼å¼çµ±ä¸€
                    content = str(index)+ '.' + con + '\n'
                    # å°‡ç•™è¨€å­—ä¸²æ•´ç†æ–°å¢åˆ°å–®ç¯‡æ–‡ç« å…§æ–‡ä¸­
                    articleTxt += content
            # å¿½ç•¥åˆªé™¤çš„ç•™è¨€æ¨“å±¤
            except:
                pass
            print('====')
            # åŸ·è¡Œå®Œä¸€æ¬¡è¿´åœˆè¡¨ç¤ºæˆ‘å€‘è¨ªå•äº†ä¸€æ¬¡çš„å–®ç¯‡æ–‡ç« å›ºå®šæ•¸é‡çš„ç•™è¨€å€ï¼Œå› æ­¤é€²è¡Œç­‰å¾…
            time.sleep(delay)
            # after éš¨è‘—æˆ‘å€‘è¨­å®šæ¯æ¬¡çˆ¬å–çš„ç•™è¨€æ¨“å±¤å¢åŠ 
            after += 30
        # è‹¥æˆ‘å€‘è¨ªå•å–®ç¯‡æ–‡ç« ç•™è¨€å€çš„ç•™è¨€ç‚ºé›¶ï¼Œå‰‡å„²å­˜æˆ‘å€‘å–å¾—çš„å…§æ–‡
        else :
            # å„²å­˜ä¸¦å¯«å…¥æˆ‘å€‘çš„å…§å®¹ï¼Œtitle æ˜¯æ–‡ç« æ¨™é¡Œï¼Œå…§å®¹æ˜¯å–®ç¯‡å…§æ–‡åŠ ä¸Šæ‰€æœ‰ç•™è¨€
            with open(r'%s\\\%s.txt' % (path, title), 'w', encoding='utf-8') as w:
                w.write(articleTxt)
            break
# print(articleTxt)
```
